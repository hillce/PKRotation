```python
import os
import csv
import scipy
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from iSCAMS_class import iSCAMS
from astroML.plotting import hist
```

```python
Path = r"./Contrasts/DRP1/"
Filenames = []
for filename in os.listdir(Path):
    filename_sample = os.path.join(os.path.expanduser(Path),filename)
    assert os.path.exists(filename_sample)
    Filenames.append(filename_sample)
Filenames = Filenames[:-1]
print("Standard Files:",Filenames)
```

```python
Conc = ["100nM","10nM","200nM","20nM","30nM","40nM","50nM"]
n = np.arange(len(Filenames))
variables = []
for x,i in zip(Conc,n):
    Cf = np.load(Filenames[i])
    globals()['DRP1_%s' % x] = iSCAMS(Cf,Conc=x,Protein="DRP1",Buffer="HEPES",Mass=False,c_range=(0.0,0.04))
    variables.append('DRP1_%s' % x)
    
```

```python
for var in variables:
    globals()[var].Auto_Gauss()
    globals()[var].Fit_Gaussian()
```

```python
for var in variables:
    globals()[var].GaussKernel(sigma=30)
    plt.figure()
    n,bins,p = hist(globals()[var].contrast_smooth,bins=100,)
    plt.clf()
    plt.plot(bins[2:-3],n[2:-2])
    plt.xlim(0.0,0.015)
    plt.show()
```
